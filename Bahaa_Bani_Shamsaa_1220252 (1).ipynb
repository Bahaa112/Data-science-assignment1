```python
df['currency_rate'] = df['price'] / df['usdprice']
df
```

This is the exchange rate between the local currency and the US dollar (USD)

```python
df['unified_price'] = None
df['unified_price'] = df.apply(
    lambda row: row['usdprice'] if 'KG' in row['unit'] or 'kg' in row['unit'] or 'L' in row['unit'] or 'l' in row['unit']
    else row['usdprice'] * 1000 if 'g' in row['unit']
    else row['usdprice'] * 1000 if 'ml' in row['unit']
    else None, axis=1)
df
```

As we see here  unified_price column  represents the price in USD per kilogram for commodities measured by weight and the price per liter for commodities measured by volume

```python
weight_units = ['KG', 'kg', 'g']
weight_commodities = df[df['unit'].isin(weight_units)]
max_weight_index = weight_commodities['unified_price'].idxmax()
max_weight_price = weight_commodities.loc[max_weight_index]
max_weight_price_info = {max_weight_price['commodity'],max_weight_price['unified_price']}
```

```python
max_weight_price_info
```

Max price for commodities measured by weight is Meat(goat,with bones) with price of 17.3144

```python
volume_units = ['L', 'l', 'ml']
volume_commodities = df[df['unit'].isin(volume_units)]
max_volume_index = volume_commodities['unified_price'].idxmax()
max_volume_price = volume_commodities.loc[max_volume_index]
max_volume_price_info = {max_volume_price['commodity'],max_volume_price['unified_price']
}
```

```python
max_volume_price_info
```

Max price for commodities measured by volume is Fuel (petrol-gasoline) with price of 2.0459
